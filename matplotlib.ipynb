plt.subplot() automatically call the add_axes method based on the nrows and ncols

```python
fig, axes = plt.subplots(nrows = 1, ncols = 2)

# axes[0] grabs the first subplot object

axes[0].plot(x, y)
axes[0].set_title('First Plot')

axes[1].plot(y, x)
axes[1].set_title('Second Plot')


plt.tight_layout()
```

<h3> Figure Size and DPI </h3>

```python
fig = plt.figure(figsize = (3, 2))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)

# figsize is a tuple which contains width and height of the figure in inches
# dpi is dots per inch or pixels per inch
```

```python
fig = plt.figure(figsize = (8, 2))

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y)

# figsize is a tuple which contains width and height of the figure in inches
# dpi is dots per inch or pixels per inch
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (8, 2))

axes[0].plot(x, y)

axes[1].plot(y, x)

plt.tight_layout()
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label = 'X Squared')     # label argument is used by the legend method
ax.plot(x, x**3, label = 'X Cubed')

ax.legend(loc = 0)      # checks each of the plot calls and see if there is a label
# loc argument specifies where to show the labels on the figure, it is linked to the location of the label string
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label = 'X Squared')     # label argument is used by the legend method
ax.plot(x, x**3, label = 'X Cubed')

ax.legend(loc = 10)      # checks each of the plot calls and see if there is a label
# loc argument specifies where to show the labels on the figure, it is linked to the location of the label string
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, x**2, label = 'X Squared')     # label argument is used by the legend method
ax.plot(x, x**3, label = 'X Cubed')

ax.legend(loc = (0.1, 0.1))      

# checks each of the plot calls and see if there is a label
# loc argument specifies where to show the labels on the figure, it is linked to the location of the label string

# for loc argument, we can specify a legend as the location of the left bottom corner of the legend wrt to axes
```
