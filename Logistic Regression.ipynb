One way to fill the missing age data is by filling in the mean age of all the passengers. This is known as imputation.

<br>

We can also fill by average age per passenger class

```python
plt.figure(figsize = (10, 7))

sns.boxplot(x = 'Pclass', y = 'Age', data = train)
```

From above plot, passengers(wealthier passengers) in class 1 and 2 tend to actually be a bit older than the passengers in the third class.

```python
def impute_age(cols):
    
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:            
            return 37
        
        elif Pclass == 2:           
            return 29
        
        else:           
            return 24
        
        
    else:
        return Age
```

```python
train['Age'] = train[['Age', 'Pclass']].apply(impute_age, axis = 1) # axis = 1 as we want to apply across the columns
```

Converting categorical features to numerical features known as dummy or indicator variables.

```python
pd.get_dummies(train['Sex'])          # we get a column for every single category in the Sex feature
```

one slight issue with the result set and that is one column is perfect predictor of the other column meaning if ML algorithm
gets fed from both cols. then algorithm will immediately know that if it is 0 female than it is going to be 1 male and this
issue is known as multi-collinearity and it basically will mess up the algo. because a bunch of calls will be perfect 
predictors of another column. In order to avoid this, we add the argument drop_first = True, this will drop the first column

```python
pd.get_dummies(train['Sex'], drop_first = True) 
```

```python
sex = pd.get_dummies(train['Sex'], drop_first = True) 
```

```python
sex.head()
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first = True) 
```

```python
embark.head() # any call is not perfect predictor of the other columns as we dropped the C column
```

```python
train = pd.concat([train, sex, embark], axis = 1)
```

```python
train.head()
```

```python
train.drop(['Sex', 'Embarked', 'Name', 'Ticket'], axis = 1, inplace = True)
```

```python
train.drop('PassengerId', axis = 1, inplace = True)
```

```python
train.head()
```

```python
X = train.drop('Survived', axis = 1)
y = train['Survived']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
print(confusion_matrix(y_test, predictions))
```
