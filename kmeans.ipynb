```python
# import relevant libraries

import seaborn as sns
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# make_blobs is used to generate some artificial data

from sklearn.datasets import make_blobs
```

```python
# n_samples refers to the number of samples to generate
# n_features refers to the number of features to generate
# centers refer to the number of cluster in the data
# cluster_std refers to the std deviation in clusters

data = make_blobs(n_samples = 200, n_features = 2, centers = 4, cluster_std=1.8, random_state=101)
```

```python
data   

# tuple -> 1st elements contains the actual data i.e 200 rows each having values for 2 features/cols

# 2nd element would refer to the actual label(cluster label) the data point belongs to
```

```python
data[0]  # numpy array -> no of samples and then two columns of features
```

```python
data[0].shape
```

```python
# we are plotting all the rows on the data(first element of tuple) with feature 1 on one axis and the other feature on the other 
# axis

plt.scatter(data[0][:, 0], data[0][:, 1])
```

```python
data[1]           # contains the cluster it belongs to, we have 0 1 2 3 -> 4 centers to these blobs
```

```python
# we are plotting all the rows on the data(first element of tuple) with feature 1 on one axis and the other feature on the other 
# axis, c = data[1] colors the data points(in scatter plot) based on the values in data[1] and we can see there are 4 clusters

plt.scatter(data[0][:, 0], data[0][:, 1], c = data[1])
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c = data[1], cmap = 'rainbow')
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
# subplot of 1 row 2 cols, sharey True so that they share that axis, then we use tuple unpacking to grab those two axes

fig, (ax1, ax2) = plt.subplots(1, 2, sharey = True, figsize = (10, 6))    

# first set of axis is going to plot original data and it's going to color that based off what the algorithm for the label 
# should look like
ax1.set_title('K Means')
ax1.scatter(data[0][:,0], data[0][:, 1], c = kmeans.labels_, cmap = 'rainbow')

# first set of axis is also going to plot original data and it's going to color that based off the actual label for the data pts.
ax2.set_title('Original')
ax2.scatter(data[0][:,0], data[0][:, 1], c = data[1], cmap = 'rainbow')
```
